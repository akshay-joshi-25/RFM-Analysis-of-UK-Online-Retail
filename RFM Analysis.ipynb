```python
import pandas as pd
from datetime import datetime

df = pd.read_csv("C:/Users/aksha/Documents/Github Datasets/UK Online Retail Data/Dataset.csv")
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
df = df.iloc[:, :-1]

df
```

```python
df['LatestInvoiceDate'] = df[df['Quantity'] > 0].groupby('CustomerID')['InvoiceDate'].transform('max')
df
```

```python
reference_date = datetime(2011, 12, 10)

df['Recency'] = (reference_date - df['LatestInvoiceDate']).dt.days

df
```

```python
frequency = (df[df['Quantity'] > 0]
             .groupby('CustomerID')['InvoiceNo']
             .count()
             .reset_index()
             .rename(columns={'InvoiceNo': 'Frequency'}))

# Merge Frequency back to the main dataframe
df = df.merge(frequency, on='CustomerID', how='left')

df
```

```python
monetary = (df.groupby('CustomerID')['TotalPrice']
             .sum()
             .reset_index()
             .rename(columns={'TotalPrice': 'Monetary'}))

# Merge Monetary back to the main dataframe
df = df.merge(monetary, on='CustomerID', how='left')

df
```

```python
rfm = pd.read_csv("C:/Users/aksha/Documents/Github Datasets/UK Online Retail Data/RMF_Dataset.csv")
rfm
```

```python
# Recency: lower is better so scoring is INVERTED (label 5 goes to lowest values)
rfm['R_Score'] = pd.qcut(rfm['Recency'], q=5, labels=[5, 4, 3, 2, 1])

# Frequency: higher is better so scoring is NORMAL
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), q=5, labels=[1, 2, 3, 4, 5])

# Monetary: higher is better so scoring is NORMAL
rfm['M_Score'] = pd.qcut(rfm['Monetary'], q=5, labels=[1, 2, 3, 4, 5])

# Combine into a single RFM score string e.g. "5-4-3"
rfm['RFM_Segment'] = (rfm['R_Score'].astype(str) + '-' +
                      rfm['F_Score'].astype(str) + '-' +
                      rfm['M_Score'].astype(str))

# Composite numeric score (sum, range 3–15)
rfm['RFM_Score'] = (rfm['R_Score'].astype(int) +
                    rfm['F_Score'].astype(int) +
                    rfm['M_Score'].astype(int))
```

```python
rfm
```

```python
# Cast columns ONCE before applying
rfm['R_Score'] = rfm['R_Score'].astype(int)
rfm['F_Score'] = rfm['F_Score'].astype(int)
rfm['M_Score'] = rfm['M_Score'].astype(int)
rfm['FM_Score'] = ((rfm['F_Score'] + rfm['M_Score']) / 2).round().astype(int).clip(1, 5)

def assign_segment(r, f, m):

    # Champions — recently active, highly frequent, high spenders
    if r >= 4 and f >= 4 and m >= 4:
        return 'Champions'

    # Can't Lose Them — were high value but have gone very quiet
    # Check EARLY before other low-recency segments absorb them
    elif r <= 2 and f >= 4 and m >= 4:
        return "Can't Lose Them"

    # At Risk — used to buy reasonably well but recency is poor
    elif r <= 2 and f >= 3 and m >= 3:
        return 'At Risk'

    # Loyal Customers — frequent and high spend, still somewhat recent
    elif r >= 3 and f >= 3 and m >= 3:
        return 'Loyal Customers'

    # New Customers — very recent but only bought once or twice
    elif r >= 4 and f <= 2:
        return 'New Customers'

    # Promising — recent, some frequency, but spend not yet high
    elif r >= 3 and f >= 2 and m <= 3:
        return 'Promising'

    # Potential Loyalist — recent, low frequency, moderate spend
    elif r >= 3 and f <= 2 and m >= 2:
        return 'Potential Loyalist'

    # Need Attention — mid recency, frequency and spend declining
    elif r == 3 and f <= 3 and m <= 3:
        return 'Need Attention'

    # About to Sleep — recency fading, low engagement across the board
    elif r == 2 and f <= 3 and m <= 3:
        return 'About to Sleep'

    # Hibernating — everything low, effectively churned
    else:
        return 'Hibernating'

# Now x['R_Score'] is already an int scalar — no .astype() needed
rfm['Segment'] = rfm.apply(
    lambda x: assign_segment(x['R_Score'], x['F_Score'], x['M_Score']),
    axis=1
)
```

```python
# Count customers per segment for annotation
segment_counts = rfm['Segment'].value_counts()
print(segment_counts)
```
